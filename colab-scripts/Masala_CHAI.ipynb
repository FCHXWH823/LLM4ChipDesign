
**Following operations use colab terminal**:
1. ```bash
   source ~/.bashrc
   ```

2. **Create a Conda environment:**
   change lines 321, 322:
   - opencv-python==4.10
   - opencv-python-headless==4.10
   
   ```bash
   conda env create -f environment.yml
   ```

3. **Activate the Conda environment:**

	```bash
   conda activate autospice_env
   ```

4. **Download hough scripts:**

    Folder containing scripts to use Hough Transform for net detection. Due to large size of the files, download the hough from this Google Drive link: https://drive.google.com/file/d/1mTwWWSMsYwhJW-GfKKVm21Lm5lVtMJ1y/view?usp=sharing
    
    and place it under Masala-CHAI.

5. **Edit path of pretrained YOLO model:**

    Line 110 of ./utils/yolov8.py: /your/path/MASALA-CHAI/trained_checkpoints/yolov8_best.pt

6. **Add sample images:**

	Place your sample images in the `./sample-images/` folder.

7. **Run the pipeline:**

	```bash
	cd Masala-CHAI & python run.py --src ./sample-images/ --tgt ./sample-output --api_key <openai_api_key>
  ```
	where:
	- `--src` : Directory path to the sample images.
	- `--tgt` : Output directory path for the generated netlists.
	- `--api_key` : Your OpenAI API key for using GPT-4
